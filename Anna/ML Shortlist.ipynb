```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
data = pd.read_csv(os.path.join('Invistico_Airline.csv'))
data
```

```python
data2 = data.dropna(how='any')
data2
```

```python
# Assign X (data) and y (target)
X = data2.drop("satisfaction", axis=1)
y = data2["satisfaction"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

length = len(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=length)
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.coef_
```

```python
feature_names = list(X.columns)
```

```python
pip install eli5
```

```python

import eli5
eli5.show_weights(classifier, feature_names = feature_names)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

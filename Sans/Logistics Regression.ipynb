```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Invistico_Airline.csv')
df.head()
```

```python
target = df["satisfaction"]
target_names = ["satisfied", "dissatisfied"]
```

```python
data = df.drop(["satisfaction"], axis=1)

data.head()
```

```python
data = pd.get_dummies(data, drop_first=True)
feature_names = data.columns
data.head(20)
```

```python
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X_train)
X_train = imputer.transform(X_train)
```

```python
X_test = imputer.transform(X_test)
```

```python
import pickle
Pkl_Filename = "imputer.pkl"  

with open(Pkl_Filename, 'wb') as file:  
    pickle.dump(imputer, file)
```

```python
X_test
```

```python
feature_names
lst = list(feature_names)
lst
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=4000000)
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
# get importance
importance = classifier.coef_
importance
```

```python
sorted(zip(importance, feature_names), reverse=True)
```
